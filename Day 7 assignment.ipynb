```python
# outliers treatment
attr.describe()
```

```python
Q1 = attr.quantile(0.25)
Q3 = attr.quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
```

```python
attr = attr[~((attr < lower) | (attr > upper)).any(axis = 1)]
```
